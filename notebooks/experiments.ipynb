```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
experiment_name = "MHIST1000-0.0.1-1673017522"
```

```python
def plot_experiment(experiment_name: str):
    loss_df = pd.read_csv(f"../experiments/{experiment_name}/loss.csv")
    plot_classification_report(experiment_name=experiment_name)
    plot_loss(loss_df=loss_df, experiment_name=experiment_name)

def plot_classification_report(experiment_name):
    with open(f"../experiments/{experiment_name}/classification_report.txt", "r") as cfile:
        print(cfile.read())

def plot_loss(loss_df, experiment_name):
    fig, ax = plt.subplots(dpi=120)
    loss_df.plot(ax=ax)
    plt.xlabel("Epoch")
    plt.ylabel("BCE Loss")
    plt.title(experiment_name)
    plt.show()
```

```python
plot_experiment(experiment_name)
```
